```python
data['sc'].fillna(data['sc'].mean(),inplace=True)
data['bu'].fillna(data['bu'].mean(),inplace=True)

data['bgr'].fillna(data['bgr'].mean(),inplace=True)
data['al'].fillna(data['al'].mean(),inplace=True)
data['sod'].fillna(data['sod'].mean(),inplace=True)
data['pot'].fillna(data['pot'].mean(),inplace=True)
data['hemo'].fillna(data['hemo'].mean(),inplace=True)

data['su'].fillna(data['su'].mean(),inplace=True)
data['age'].fillna(data['age'].mean(),inplace=True)
data['bp'].fillna(data['bp'].mean(),inplace=True)
data['sg'].fillna(data['sg'].mean(),inplace=True)


```

```python
Numcols.remove('pcv')
Numcols.remove('rbc')
Numcols.remove('wc')
Catcols.add('pcv')
Catcols.add('rbc')
Catcols.add('wc')
print(Catcols)
print(Numcols)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in Catcols:
    print("LABEL ENCODING OF :",i)
    LEi = LabelEncoder()
    print(c(data[i]))
    data[i] = LEi.fit_transform(data[i])
    print(c(data[i]))
    print("*"*100)
```

```python
#rbc,pc,bgr,bu,pe,ane,dm,cad
selcols = ['rbc','pc','bgr','bu','pe','ane','dm','cad']
x = pd.DataFrame(data,columns = selcols)
y = pd.DataFrame(data, columns = ['classification'])

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 2)


print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression()
lgr.fit(x_train,y_train)

```

```python
y_pred = lgr.predict(x_test)
```

```python
y_pred1 = lgr.predict([[129,99,1,0,0,1,0,1]])
print(y_pred)
c(y_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
conf_mat =  confusion_matrix(y_test,y_pred)
print(conf_mat)
```
